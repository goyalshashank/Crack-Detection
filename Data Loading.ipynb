### Imports 

```python
from PIL import Image
from matplotlib.pyplot import imshow
import pandas
import matplotlib.pylab as plt
import os
import glob
```

```python
def show_data(data_sample, shape = (28, 28)):
    plt.imshow(data_sample[0].numpy().reshape(shape),cmap='gray')
    plt.title('y= '+data_sample[1])
```

```python
directory = ""
negative = "Negative"
positive = "Positive"

negative_file_path = os.path.join(directory,negative)
positive_file_path = os.path.join(directory, positive)

positive_file_path
```

### Loading File Path of Each Image
Now Lets print first three elements of the list using listdir method.

```python
os.listdir(negative_file_path)[0:3]
```

##### For full path of image we need to join the file path with file name

```python
negative_files = [os.path.join(negative_file_path,file) for file in os.listdir(negative_file_path) if file.endswith(".jpg")]
positive_files = [os.path.join(positive_file_path, file) for file in os.listdir(positive_file_path) if file.endswith(".jpg")]

negative_files.sort()
positive_files.sort()

positive_files[0:3]
```

### Display and Analyze Image

```python
image1 = Image.open(negative_files[0])
image2 = Image.open(positive_files[0])

image1
```
